```python
import pandas as pd
import random
from konlpy.tag import Okt  # 명사 추출
```

```python
lets_study = pd.read_excel(r'dummies.xlsx')
```

```python
lets_study
```

```python
# get category
categories = ['r','python','c','java']
places = ['강남역', '역삼역', '교대역']
category = []
place = []
for i in range(lets_study.shape[0]):
    category.append(random.choice(categories))
    place.append(random.choice(places))
lets_study['category'] = category
lets_study['place'] = place
lets_study['user'] = 1
```

```python
# tokenize
okt = Okt()
lets_study['content_token'] = lets_study['content'].apply(lambda content: ','.join(okt.nouns(content)))
```

```python
lets_study
```

```python
lets_study['content_token']=lets_study['content_token'].apply(lambda x: '빈칸' if len(x)==0 else x)
```

```python
lets_study.to_excel('lets_study.xlsx', index=False)
```
