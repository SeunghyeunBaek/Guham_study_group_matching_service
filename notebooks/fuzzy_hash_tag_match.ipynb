```python
from random import *
import pandas as pd
import numpy as np
from pandas import DataFrame
%matplotlib inline
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
# generate 100 dummy data
hash_tags = ['r', 'python', 'java', 'c', 'ruby']
tmp_dict = {
    'user_id':[],
    'hash_tags':[]
}

# 1 dummy has 1-5 tags
for i in range(100):
    n_tags = randint(1,5)
    random_tags = sample(hash_tags, n_tags)
    tmp_dict['user_id'].append(i)
    tmp_dict['hash_tags'].append(random_tags)

tmp_df = pd.DataFrame.from_dict(tmp_dict)
tmp_df.head(10)
```

```python
def sort_and_make_string(hash_tag_list):
    hash_tag_list.sort()
    return ' '.join(hash_tag_list)
    
    
def get_ratio(my_hash_tags, ur_hash_tags):
    # 나와 상대방의 해시태그를 알파벳 순으로 정렬한다.
    # 나와 상대방의 해시태그를 스트링으로 변환한다.
    my_hash_str = sort_and_make_string(my_hash_tags)
    ur_hash_str = sort_and_make_string(ur_hash_tags)
    # fuzzy match ratio 반환
    return fuzz.ratio(my_hash_str, ur_hash_str)
```

```python
my_tag = ['r', 'python']
tmp_df['fuzzy_ratio'] = tmp_df['hash_tags'].apply(lambda ur_tag: get_ratio(my_tag, ur_tag))
```

```python
# fuzzy_ratio 순으로 내림차순 정렬
tmp_df.sort_values(by=['fuzzy_ratio'], ascending=False)
```
